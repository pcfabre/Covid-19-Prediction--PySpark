```
filename = "/df_for_dashboard1.csv"
```

```
df_for_dashboard = spark.read.format("csv").option("header","True").option("inferSchema","True").load(filename)
```

```
import pyspark.sql.functions as psf
```

```
df_for_dashboard = df_for_dashboard.withColumn("country", psf.trim(df_for_dashboard["country"]))
```

```
display(df_for_dashboard)
```

```
file_country_codes = "/FileStore/tables/wikipedia_iso_country_codes-13d5f.csv"
iso_country_codes = spark.read.format("csv").option("header","True").option("inferSchema","True").load(file_country_codes)
```

```
iso_country_codes = iso_country_codes.withColumn("country", psf.col("English short name lower case"))
```

```
display(iso_country_codes)
```

```
df_for_dashboard = df_for_dashboard.join(iso_country_codes.select("country", "Alpha-3 code"), df_for_dashboard.country == iso_country_codes.country)
```

```
df_for_dashboard = df_for_dashboard.withColumnRenamed("Alpha-3 code", "Alpha3_code")
```

```
display(df_for_dashboard)
```

```
df_for_dashboard.printSchema()
```

```
displayHTML("""<font size="6" color="red" face="sans-serif">Dashboard of COVID-19 evolution</font>""")
```

```
df_for_dashboard.registerTempTable("df")
```

```
display(spark.sql("SELECT reporting_date AS date, count(id) AS Nb_Hospitalisations FROM df WHERE visited_hospital = 1 GROUP BY reporting_date"))
```

```
display(spark.sql("SELECT Alpha3_code, count(id) FROM df GROUP BY Alpha3_code"))
```

```
display(spark.sql("SELECT reporting_date, state, count(id) FROM df GROUP BY reporting_date, state ORDER BY reporting_date"))
```
